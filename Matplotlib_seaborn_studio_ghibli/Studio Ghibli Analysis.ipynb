```python
# function to convert string ghibli_df["Duration"] to minutes int

def time_to_minutes (time_str):
    parts = time_str.split()
    
    hours =0
    minutes = 0
    
    #deletes the letters and converts the string to integers
    for part in parts:
        if part.endswith('h'):
            hours=int(part[:-1])
        elif part.endswith('m'):
            minutes=int(part[:-1])
    
    #multiplies the hour by sixty and adds the minutes to get the full time in minutes
    total_minutes= hours * 60 + minutes
    
    return total_minutes
```

```python
#Applying the function to the column
ghibli_df["Duration"] = ghibli_df['Duration'].apply(time_to_minutes)
```

```python
ghibli_df
```

```python
#renaming the duration column with unit
ghibli_df.rename(columns={"Duration": "Duration (m)"})
```

```python
#dropping genre 3 because it has null values
ghibli_df.drop('Genre 3', axis=1, inplace=True)
```

```python
ghibli_df
```

```python
ghibli_df["Name"] = ghibli_df["Name"].replace("\n", '', regex=True)
```

```python
ghibli_df
```

```python
#eliminating the date at the end of some of the names
ghibli_df["Name"] = ghibli_df["Name"].str.replace(r'\s*\(\d{4}\)$', '', regex=True)
```

```python
#strip dollar signs
ghibli_df["Budget"] = ghibli_df["Budget"].str.strip("$")
```

```python
#strip dollar signs
ghibli_df["Revenue"] = ghibli_df["Revenue"].str.strip("$")
```

```python
ghibli_df["Budget"] = ghibli_df["Budget"].str.replace(r'\.\d*', '', regex=True)
```

```python
ghibli_df["Revenue"] = ghibli_df["Revenue"].str.replace(r'\.\d*', '', regex=True)
```
